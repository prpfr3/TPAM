```python
wiki_wiki = wikipediaapi.Wikipedia(
        language='en',
        extract_format=wikipediaapi.ExtractFormat.HTML
)

p_wiki = wiki_wiki.page('LMS_Stanier_Class_8F')
print(p_wiki.text)
```

```python
def print_categorymembers(categorymembers, level=0, max_level=1):
        for c in categorymembers.values():
            print("%s: %s (ns: %d)" % ("*" * (level + 1), c.title, c.ns))
            if c.ns == wikipediaapi.Namespace.CATEGORY and level < max_level:
                print_categorymembers(c.categorymembers, level=level + 1, max_level=max_level)


cat = wiki_wiki.page("Category:Railway lines in the United Kingdom")
print("Category members: Category:Railway lines in the United Kingdom")
print_categorymembers(cat.categorymembers)
```

```python
import requests, csv, os
from bs4 import BeautifulSoup
url = "https://en.wikipedia.org/wiki/Glossop_line"
try:
    res = requests.get(url)
    res.raise_for_status()
except requests.exceptions.ConnectionError as err:
    # eg, no internet
    raise SystemExit(err)
except requests.exceptions.HTTPError as err:
    # eg, url, server and other errors
    raise SystemExit(err)
soup = BeautifulSoup(res.text, 'html.parser') 
```
